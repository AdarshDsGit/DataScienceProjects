```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data  = pd.read_excel('Data_Mendeley.xlsx')
```

```python
data.drop(['Ent', 'LHundreds','LFours','LRuns','LAve','LStrRate','LFifties','LSixes','LCatches','LStumps','LWkts','LEcon','LFourWkts','LFiveWkts'],axis=1,inplace=True)
data.columns
```

```python
data.info()
```

```python
data['Name'].unique()
```

```python
top_scored_player = data.sort_values(['Runs'],ascending=False)
top_scored_player
```

```python
sns.barplot(x=top_scored_player['Name'][:10],y=top_scored_player['Runs'][:10],palette='magma')
plt.title('Players With Top Runs ', fontweight='bold',fontsize=15)
```

```python
Highest_sixes = data.sort_values(['Sixes'],ascending=False)
Highest_sixes
```

```python
sns.barplot(x=Highest_sixes['Name'][:10],y=Highest_sixes['Sixes'][:10],palette='magma')
plt.title('Players with Highest Sixes', fontweight='bold',fontsize=15)
```

```python
data['Nationality'].unique()
```

```python
df1 = data.groupby('Team')['Runs'].sum()
df1
```

```python
plt.figure(figsize=(15,10))
df1.sort_values(ascending=False).plot(kind='bar')
plt.title('Team Runs', fontweight='bold',fontsize=15)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [17,10]
Sixes = data[['Sixes', 'Matches', 'Team']].groupby(['Team']).mean().sort_values(by='Matches',ascending=False)
display(Sixes)
Sixes.plot(kind='bar')
plt.title('Teams Sixes and Matches', fontweight='bold',fontsize=15)
plt.show()
```

```python
centuary = data.sort_values(['Hundreds'],ascending=False)
centuary
```

```python
fig,ax = plt.subplots(figsize=(20,10))
ax.set(facecolor='black')
sns.barplot(x=centuary['Team'],y=centuary['Hundreds'],palette='hls')
plt.title('Top Teams', fontweight='bold',fontsize=15)
plt.xlabel('Teams', fontsize=15,fontweight='bold')
plt.ylabel('Hundreds',fontsize=15,fontweight='bold')
plt.show()
```

```python
df2 = data.groupby('Name')['Final Price'].sum()
df2
```

```python
top_paid = data.sort_values(['Final Price'],ascending=False)
top_paid
```

```python
fig,ax = plt.subplots(figsize=(15,5))
ax.set(facecolor='black')
sns.barplot(x=top_paid['Name'][:12],y=top_paid['Final Price'][:12],palette='hls')
plt.title('7 Top Paid Players', fontweight='bold',fontsize=15)
plt.xlabel('Name', fontsize=15,fontweight='bold')
plt.ylabel('Paid',fontsize=15,fontweight='bold')
plt.show()
```
