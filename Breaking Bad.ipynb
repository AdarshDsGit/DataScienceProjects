```python
plt.figure(figsize=(12,7))
plt.style.use('seaborn')
sns.lineplot(x='Date',y='U.S. viewers_million',data=bb)
```

```python
plt.figure(figsize=(12,7))
sns.countplot(bb['Directed by'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.style.use('fivethirtyeight')
sns.lineplot(x='Date',y='Rating_IMDB',data=bb)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.style.use('ggplot')
sns.lineplot(x='Title',y='Rating_IMDB',data=bb)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.style.use('ggplot')
sns.lineplot(x='Written by',y='Rating_IMDB',data=bb)
plt.xticks(rotation=90)
plt.show()
```
