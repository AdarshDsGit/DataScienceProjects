```python
df = pd.read_excel('Country-Code.xlsx')
final_data = pd.merge(data,df,on='Country Code',how='left')
```

```python
final_data.head()
```

```python
final_data.columns
```

```python
country_names = final_data.Country.value_counts().index
country_vlaues  = final_data.Country.value_counts().values
```

```python
plt.pie(x=country_vlaues[:3],labels=country_names[:3],autopct="%1.2f%%")
plt.title('Top 3 Country')
```

Zomato's major part of business is held in India then comes United States and United Kingdom

```python
final_data.columns
```

```python
ratings = final_data.groupby(['Aggregate rating', 'Rating color','Rating text' ]).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

- from 4.5 to 4.9 ---Exellent rating
- from 4.0 to 4.4 ---Very Good
- from 3.5 to 3.9 ---Good
- from 2.5 to 3.4 ---Average
- from 1.8 to 2.4 ---Poor
- 0 ---Not Rated

```python
matplotlib.rcParams['figure.figsize'] = (15,10)
sns.set_style('dark')
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['blue', 'red','orange','yellow','green','green'])
```

- Not Rated Count is High
- Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue', 'red','orange','yellow','green','green'])
```
